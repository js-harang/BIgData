```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'NanumGothic'

file_path = "/content/data/MyDrive/서울특별시 공공자전거 이용정보(일별)_2306.csv"
bike_data = pd.read_csv(file_path, encoding='cp949')

selected_columns = ['대여일자', '대여소번호', '성별', '연령대', ' 이용건수']
bike_data = bike_data[selected_columns]

bike_data.dropna(inplace=True)

bike_data['대여일자'] = pd.to_datetime(bike_data['대여일자'], format='%Y-%m-%d')

plt.figure(figsize=(12, 8))
sns.scatterplot(x='대여일자', y=' 이용건수', hue='성별', style='연령대', data=bike_data)
plt.title('대여일자별 이용건수에 따른 성별 및 연령대 분포')
plt.show()
```
